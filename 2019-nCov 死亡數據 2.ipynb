```python
import numpy as np
import io
import requests
import pandas as pd
url="https://www.chp.gov.hk/files/misc/latest_situation_of_reported_cases_covid_19_eng.csv"
s=requests.get(url).content
df=pd.read_csv(io.StringIO(s.decode('utf-8')))
df=df[['As of date','Number of death cases']]
```

```python
import matplotlib.pyplot as plt
plt.plot(df['As of date'],df['Number of death cases'])
```

```python
df[df['Hospitalised/Discharged/Deceased']=="Deceased"].count()['Report date']
```

```python
df =pd.pivot_table(df,values='Case no.',index='Report date',columns='Hospitalised/Discharged/Deceased',aggfunc='count')
```
